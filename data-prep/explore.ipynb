```python
import geopandas as gpd
```

```python
NE_SUBUNITS = "data/ne/ne_10m_map_subunits_admin_0_countries.geojson"
NE = "data/ne/ne_10m_admin_0_countries.geojson"
```

```python
gdf = gpd.read_file(NE_SUBUNITS)
```

```python
len(gdf['SOVEREIGNT'].unique())
```

```python
len(gdf['SUBUNIT'].unique())
```

```python
# iterate over all the unique sovereign states and print their subunits
# only print sovereigns with more than one subunit
for sovereign in gdf['SOVEREIGNT'].unique():
    subunits = gdf[gdf['SOVEREIGNT'] == sovereign]['SUBUNIT'].unique()
    if len(subunits) > 1:
        print(f"{sovereign}: {subunits}")
```
